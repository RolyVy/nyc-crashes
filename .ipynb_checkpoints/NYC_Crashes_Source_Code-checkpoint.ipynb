# Working on the addresses/location of the accidents

```python
# Replacing NaN values by empty strings before turning the streets columns values into string 
df= df.replace(np.nan, '', regex=True)
```

```python
#Concatenating the columns. As it was as object type i changed the type to string to be able to concatenate the values
df["street_name"] = df["on_street_name"].astype(str) + df["off_street_name"].astype(str) + df["cross_street_name"].astype(str)
```

```python
#Verifying no more NaN values
df["street_name"].isnull().sum()
```

```python
df["street_name"]
```

```python
#Verifying if there are still rows without streets
df.loc[df['street_name']=='']
```

```python
#Getting the index of the 42 rows missing a street
index_rows_with_no_street_name=df[df['street_name'] == ""].index.tolist()
```

```python
index_rows_with_no_street_name
```

```python
len(df)
```

```python
#Dropping the 42 rows missing a street
df.drop(index_rows_with_no_street_name)
```

```python
df = df.drop(index_rows_with_no_street_name)
```

```python
len(df)
```

```python
#dropping the 3 initial street columns to only keep the newly created 'street_name'


df.drop(["on_street_name","off_street_name","cross_street_name"], axis=1)
```

```python
df = df.drop(["on_street_name","off_street_name","cross_street_name"], axis=1)
```
